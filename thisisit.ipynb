```python
Y=train_data["treatment"]
X=train_data.drop(['treatment'], axis=1)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.001,random_state=0)
```

```python
from sklearn.metrics import mean_squared_error
from sklearn.ensemble import RandomForestClassifier
regressor=RandomForestClassifier(n_estimators=10000,criterion = 'entropy',random_state=0)
regressor.fit(X_train,Y_train)
```

```python
Y_pred=regressor.predict(X_test)
mean_squared_error(Y_test,Y_pred)
```

```python
from sklearn.ensemble import ExtraTreesClassifier
regressor=ExtraTreesClassifier(n_estimators=10000,max_features= 100,criterion= 'entropy',min_samples_split= 2,max_depth= 50, min_samples_leaf= 5)
regressor.fit(X_train,Y_train)
```

```python
Y_pred=regressor.predict(X_test)
mean_squared_error(Y_test,Y_pred)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(Y_test,Y_pred)
```
